```python
file_name='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
HouseDF=pd.read_csv(file_name)
HouseDF.drop(["id", "Unnamed: 0"], axis = 1, inplace = True)
HouseDF.describe()
```

```python
HouseDF.columns
```

```python
sns.pairplot(HouseDF)
```

```python
sns.heatmap(HouseDF.corr(), annot = True)
```

```python
HouseDF.replace([np.inf, -np.inf], np.nan, inplace = True)
HouseDF.dropna(inplace=True)
X = HouseDF[['sqft_living']]

y = HouseDF['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state=101)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
coeffDF = pd.DataFrame(lr.coef_,X.columns,columns=['Coefficient'])
```

```python
coeffDF
```

```python
pred = lr.predict(X_test)
```

```python
plt.scatter(y_test,pred)
```

```python
sns.distplot((y_test-pred),bins=50);
```
